```sql
%python
dbutils.widgets.removeAll()
```

```sql
%python
dbutils.widgets.text("bronzeContainer","bronze")
dbutils.widgets.text("silverContainer","silver")
dbutils.widgets.text("goldenContainer","golden")
dbutils.widgets.text("nameStorage","adlssmartdada2")
dbutils.widgets.text("catalogo","catalog_smartdata_final")
```

```sql
%python

bronzeContainer = dbutils.widgets.get("bronzeContainer")
silverContainer = dbutils.widgets.get("silverContainer")
goldenContainer = dbutils.widgets.get("goldenContainer")

nameStorage = dbutils.widgets.get("nameStorage")
catalogo = dbutils.widgets.get("catalogo")

ruta_bronze = f"abfss://{bronzeContainer}@{nameStorage}.dfs.core.windows.net"
ruta_silver = f"abfss://{silverContainer}@{nameStorage}.dfs.core.windows.net"
ruta_golden = f"abfss://{goldenContainer}@{nameStorage}.dfs.core.windows.net = f"abfss://{nameContainer}@{nameStorage}.dfs.core.windows.net"
```

### Eliminacion tablas Bronze

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.bronze.ev_population_bronze;
DROP TABLE IF EXISTS ${catalogo}.bronze.ev_specs_2025_bronze;
```

```sql
%python
## REMOVE DATA (Bronze)
dbutils.fs.rm(f"{ruta_bronze}/tablas/CLIENTES", True)
dbutils.fs.rm(f"{ruta_bronze}/tablas/VENTAS", True)
```

### Eliminacion tablas Silver

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.silver.ventas_limpias;
DROP TABLE IF EXISTS ${catalogo}.silver.clientes_validos;
DROP TABLE IF EXISTS ${catalogo}.silver.detalle_productos;
```

```sql
%python
# REMOVE DATA (Silver)
dbutils.fs.rm(f"{ruta_silver}/tablas/VENTAS_LIMPIAS", True)
dbutils.fs.rm(f"{ruta_silver}/tablas/CLIENTES_VALIDOS", True)
dbutils.fs.rm(f"{ruta_silver}/tablas/DETALLE_PRODUCTOS", True)
```

### Eliminacion tablas Golden

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.golden.kpis_ventas;
DROP TABLE IF EXISTS ${catalogo}.golden.segmentos_clientes;
```

```sql
%python
# REMOVE DATA (Golden)
dbutils.fs.rm(f"{ruta_golden}/tablas/KPIS_VENTAS", True)
dbutils.fs.rm(f"{ruta_golden}/tablas/SEGMENTOS_CLIENTES", True)
```
